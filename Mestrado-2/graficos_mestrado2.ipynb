```python
import matplotlib.pyplot as plt
import squarify
import matplotlib as mpl
```

```python
#For scenary B,here are the itens and respective values
itens=['Sementes vieira', 'Elástico para fechamento das valvas', 'Mão de obra temporária (vieira)',
'Depreciação equipamentos (vieira)', 'Juvenis bijupirá', 'Ração',  'Trash fish (engorda)', 'Mão de obra temporária',
'Depreciação equipamentos (bijupira)', 'Mão de obra permanente','Combustível barco',
'Material de limpeza e manutenção', 'E.P.I', 'Monitoramento ambiental', 'Água, eletricidade e telefone',]

#COT= Total Operting Costs in Portuguese
cot=[17920, 336, 3360, 31355, 28627.20, 80156.16, 40078.08, 1680, 19460,  45609.98, 2688, 1680, 2464, 4480, 2016]
```

```python
#uniting itens in cot so we can reorder them both at the same time
zipped = zip(cot, itens)

#Sorting the itens by cot value. Reverse is true so the higher values come first
sorted_zip = sorted(zipped, reverse=True)

#unzipping them back into to lists
cot, itens = zip(*sorted_zip)
```

```python
#plotting treemap
#mpl.rcParams.keys()

#using list comprehension so the data format can bi a list
#have to use the np.arange becase the noraml range doesn't accept float
list_marks =[i for i in np.arange(0, 1, 1/len(cot))]

#get color map
#https://matplotlib.org/stable/gallery/color/colormap_reference.html
#https://seaborn.pydata.org/generated/seaborn.color_palette.html
#https://github.com/katewall/medium_tutorials/blob/main/210219_Medium_Squarify_Matplotlib.ipynb
cmap = mpl.cm.get_cmap('viridis')

#now, it's time to apply the map to the list_marks
mapped_list = [cmap(i) for i in np.arange (0, 1, 1/len(cot))]

figure, ax = plt.subplots(1, figsize = (16,12))

squarify.plot(sizes=cot, label=itens[:8], alpha=0.8, pad=True, color = mapped_list)

ax.axis('off');
```
