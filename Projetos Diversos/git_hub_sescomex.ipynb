## Objeitvo

Esse script teve como objetivo analisar os dados obtidos do site da Secretaria de Comércio Exterior a fim de contabilizar o valor monetário das exportações dos produtos relacionados à algas no Brasil em 2019. Por se tratar de um conjunto de dados relativamente grande, não era possível fazer isso pelo excel

```python
import numpy as np
import pandas as pd
```

```python
#Carregando o arquivo
df = pd.read_csv('file_path/EXP_2019.csv', sep=';')
```

```python
#Dando uma olhada rápida na "cara" do arquivo
df.head()
```

```python
#Verificando o tipo de dado de cada variável
df.dtypes
```

```python
#Selecionando as linhas referentes às exportações somente dos produtos de interesse (que estavam ainda com o código de id)
df2 = df[df['CO_NCM'].isin([12121000, 12122000, 12122100, 12122900, 13023910,13023990])]
```

```python
df2.describe()
```

```python
#Trocando o cógidos pelas palavras que elkes representam
df2['CO_NCM'].replace([12121000, 12122000, 12122100, 12122900, 13023910,13023990],
                                     ['Alfarroba', 'Algas frescas','Algas proprias', 'outras algas',
                                      'Carragenina', 'Produtos mucilaginosos'], inplace=True )
```

```python
#Checando se fiocu como eu gostaria, a despeito do aviso de cima
df2.head()
```

```python
#Somando os valores das exportações por produto
df2.groupby(by='CO_NCM').sum()
```
